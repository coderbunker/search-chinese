```python
import pandas as pd
import bs4
import urllib3
import json
```

```python
http = urllib3.PoolManager()
```

```python
r = http.request('GET', 'http://www.oiponline.org/r/v1/sites/11209330/blog?expand=blogPosts&limit=1000&page=1&exclude_content=true')
json_text = r.data.decode('utf-8')
posts = json.loads(json_text)
```

```python
urls = [ {'publicUrl': post['publicUrl'], 'title': post['title'], 'iconUrl': post['iconUrl'] } for post in posts['data']['blog']['blogPosts']]
urls
```

```python
r = http.request('GET', 'http://www.oiponline.org/blog/rudi')
r.status
html_doc = r.data.decode('utf-8')
```

```python
soup = bs4.BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
```

```python
contents = soup.select('.s-component-content > p')
content = ' '.join([s.string for s in contents])
```

```python
import json
import string

r = http.request('GET', 'https://raw.githubusercontent.com/6/stopwords-json/master/dist/zh.json')
json_text = r.data.decode('utf-8')
stopwords_cn = json.loads(json_text)

r = http.request('GET', 'https://raw.githubusercontent.com/6/stopwords-json/master/dist/en.json')
json_text = r.data.decode('utf-8')
stopwords_en = json.loads(json_text)

```

```python
import zhon.hanzi
stopwords = stopwords_cn + stopwords_en + list(string.punctuation) + list(zhon.hanzi.punctuation)
```

```python
import jieba
seg_list = jieba.cut(content, cut_all=False)
words = set([s.strip().upper() for s in seg_list if len(s.strip())>1 and s not in stopwords])
len(words)
content
```
